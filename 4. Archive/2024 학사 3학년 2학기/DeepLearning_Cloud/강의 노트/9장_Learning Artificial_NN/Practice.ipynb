```python
from sklearn import datasets
import random
import numpy as np
```

```python
def SIGMOID(x):
    return 1 / (1 + np.exp(-x))

def SLP_SGD(tr_X, tr_y, alpha, rep):
    # initialize W (tr_X.shape[1] -> 입력 차원, tr_y.shape[1] -> 출력 차원)
    n_input = tr_X.shape[1]
    n_output = tr_y.shape[1]
    random.seed = 123
    w = random.sample(range(1, 100), n_input * n_output)
    w = (np.array(w) - 50) / 100
    w = w.reshape(n_input, n_output)

    # update w
    for i in range(rep):
        for k in range(tr_X.shape[0]):
            x = tr_X[k, :]  # 입력 데이터
            v = np.matmul(x, w)  # 가중치와 입력 데이터의 행렬곱
            y = SIGMOID(v)  # 활성화 함수 적용
            e = tr_y[k, :] - y  # 오차 계산
            w = w + alpha * np.outer(x, y * (1 - y) * e)  # 외적을 사용해 차원 맞추기

        print("error", i, np.mean(e))
    return w

```

```python
## prepare dataset ############################
iris = datasets.load_iris()
X = iris.data
target = iris.target

# one hot encoding
num = np.unique(target, axis=0)
num = num.shape[0]
y = np.eye(num)[target]

## Training (get W) ##########################
W = SLP_SGD(X, y, alpha=0.01, rep=1000)
```

```python
## Test ######################################
pred = np.zeros(X.shape[0])
for i in range(X.shape[0]):
    v = np.matmul(X[i,:], W)
    y = SIGMOID(v)

    pred[i] = np.argmax(y)
    print("target, predict", target[i], pred[i])

print("accuracy :", np.mean(pred == target))
```
