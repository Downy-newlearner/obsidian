```python
import pandas as pd
# national institute of ecologt roadkill
df = pd.read_csv('./data/국립생태원_로드킬 정보시스템 로드킬 신고 현황_20201231.csv')
```

```python
# datetime으로 변환
df['접수일시'] = pd.to_datetime(df['접수일시'])

# 년도의 범위 확인
df['접수일시'].dt.year.unique()
```

```python
start_date = df['접수일시'].min()
end_date = df['접수일시'].max()

print("전체 날짜 범위: {}에서 {}까지".format(start_date, end_date))
```
