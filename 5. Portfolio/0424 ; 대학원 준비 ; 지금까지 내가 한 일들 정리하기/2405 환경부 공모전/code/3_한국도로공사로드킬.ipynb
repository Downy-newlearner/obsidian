### 본부명

- 로드킬 발생 빈도가 많은 본부: 대전충남 > 강원 > 충북
- 대체로 2019->2020에 증가, 2020->2021 감소, 2021->2022 큰 변화 없음 
    - 광주전남의 경우 2020->2021 증가한다.
    - 충북은 2021->2022 증가한다.
    - 강원은 2021->2022 증가한다.

```python
df.groupby(['본부명']).agg({'발생건수': 'sum'}).sort_values(by='발생건수')
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(df['본부명'], df['발생건수'])

# 그래프 제목과 축 레이블 설정
plt.title('본부별 발생 건수')
plt.xlabel('본부명')
plt.ylabel('발생 건수')

# 그래프 보이기
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 연도별 본부명에 따른 발생 건수 데이터프레임 생성
yearly_counts = df.groupby(['Year', '본부명']).agg({'발생건수': 'sum'}).reset_index()

# 라인 그래프 그리기
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_counts, x='Year', y='발생건수', hue='본부명', marker='o')

# 그래프 제목과 축 레이블 설정
plt.title('연도별 본부별 발생 건수 추이')
plt.xlabel('연도')
plt.ylabel('발생 건수')

# 그래프 보이기
plt.xticks(rotation=45)
plt.legend(title='본부명', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 위도, 경도

```python
import folium
from folium.plugins import HeatMap

korea = folium.Map(location=[36.55, 127.75], zoom_start=7)
HeatMap(data=df[['위도', '경도']], radius=10).add_to(korea)

korea
```

```python
import folium

# 지도의 초기 위치와 확대 수준 설정
korea = folium.Map(location=[36.55, 127.75], zoom_start=7)

# df 데이터프레임에 있는 각 위치에 작은 빨간색 마커 추가
for idx, row in df.iterrows():
    icon = folium.Icon(color='red', icon='info-sign')
    marker = folium.Marker(
        location=[row['위도'], row['경도']],
        icon=icon
    )
    marker.add_to(korea)

# 지도 표시
korea.get_root().html.add_child(folium.Element("""
<style>
    .leaflet-marker-icon {
        width: 10px;
        height: 10px;
        margin-left: -5px;
        margin-top: -10px;
    }
</style>
"""))

korea
```

```python
import folium
from folium.plugins import MarkerCluster

# 지도의 초기 위치와 확대 수준 설정
korea = folium.Map(location=[36.55, 127.75], zoom_start=7)

# 마커 클러스터 생성
mc = MarkerCluster()

# df 데이터프레임의 각 행에 대해 마커 추가
for _, row in df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['본부명'] + ' ' + row['지사명'] + ' ' + row['노선명'] + ' ' + str(row['발생건수']),  # 발생건수를 문자열로 변환
        icon=folium.Icon(color='red')  # 빨간색 마커
    ).add_to(mc)

# 마커 클러스터를 지도에 추가
korea.add_child(mc)

# 지도 표시
korea
```
