```python
import pandas as pd

df = pd.read_csv('./data/도로공사 로드킬 (2019-2022).csv')
df.head()
```

```python
df.info()
```

```python
df.groupby(['노선명', '지사명', '구간']).sum('발생건수').sort_values('발생건수', ascending=False).head(10)
```

```python
df.groupby(['노선명', '지사명', '구간', 'Year']).sum('발생건수').sort_values('발생건수', ascending=False).head(10)
```

```python
df.groupby(['노선명', '구간', 'Year']).sum('발생건수').sort_values('발생건수', ascending=False).head(10)
```

### 홍천 315~320

```python
df[(df['지사명'] == '홍천') & (df['구간'] == '315~320')]
```

```python
df[(df['지사명'] == '홍천') & (df['구간'] == '315~320')]['Year'].unique()
```

```python
hc = df[(df['지사명'] == '홍천') & (df['구간'] == '315~320')]
```

```python
hc_yr = hc.groupby('Year').sum('발생건수')
```

```python
hc_yr
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

hc_yr.plot(kind='bar', y='발생건수', title='홍천 315~320 구간 로드킬 발생건수')
plt.show()
```

```python
hc = df[(df['지사명'] == '홍천')]
```

```python
import folium
from folium.plugins import HeatMap

lat = hc['위도'].mean()
lon = hc['경도'].mean()
map_center = [lat, lon]
korea = folium.Map(location=map_center, zoom_start=8)

heat_data = [[row['위도'], row['경도'], row['발생건수']] for index, row in hc.iterrows()]

HeatMap(data=heat_data, radius=15, max_zoom=13).add_to(korea)

korea
```

### 공주 85-90

```python
gonju = df[(df['지사명'] == '공주') & ((df['구간'] == '85~90') | (df['구간'] == '70~75'))]
gonju.head()
```

```python
import folium

icon_url_red = 'https://raw.githubusercontent.com/pointhi/leaflet-color-markers/master/img/marker-icon-2x-red.png'

lat = gonju['위도'].mean()
lon = gonju['경도'].mean()
map_center = [lat, lon]
korea_map = folium.Map(location=map_center, zoom_start=13)

for _, row in gonju.iterrows():
    icon = folium.CustomIcon(
        icon_url_red,
        icon_size=(10, 15)
    )
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"구간:  {row['구간']}, 발생건수: {row['발생건수']}",
        icon=icon
    ).add_to(korea_map)

korea_map
```
