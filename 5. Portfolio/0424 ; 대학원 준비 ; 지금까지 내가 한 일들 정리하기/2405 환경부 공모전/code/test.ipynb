```python
import pandas as pd

# CSV 파일을 읽어 데이터프레임으로 변환
csv_file = './data/도로공사 로드킬 (2019-2022).csv'
df = pd.read_csv(csv_file)

# HTML 파일 생성
html_content = """
<!DOCTYPE html>
<html>
<head>
    <meta charset="utf-8">
    <meta content="IE=edge" http-equiv="X-UA-Compatible">
    <title>환경공간정보서비스 - 토지피복지도 예제</title>
    <style type="text/css">
        html, body, #map { margin: 0; padding: 0; width: 100%; height: 100%; }
        .ol-popup {
            position: absolute;
            background-color: white;
            -webkit-filter: drop-shadow(0 1px 4px rgba(0,0,0,0.2));
            filter: drop-shadow(0 1px 4px rgba(0,0,0,0.2));
            padding: 15px;
            border-radius: 10px;
            border: 1px solid #cccccc;
            bottom: 12px;
            left: -50px;
            min-width: 180px;
        }
        .ol-popup:after, .ol-popup:before {
            top: 100%;
            border: solid transparent;
            content: " ";
            height: 0;
            width: 0;
            position: absolute;
            pointer-events: none;
        }
        .ol-popup:after {
            border-top-color: white;
            border-width: 10px;
            left: 48px;
            margin-left: -10px;
        }
        .ol-popup:before {
            border-top-color: #cccccc;
            border-width: 11px;
            left: 48px;
            margin-left: -11px;
        }
    </style>
    <link rel="stylesheet" type="text/css" href="https://cdnjs.cloudflare.com/ajax/libs/ol3/3.11.1/ol.min.css">
    <script src="https://cdnjs.cloudflare.com/ajax/libs/ol3/3.11.1/ol.min.js"></script>
</head>
<body>
<div id="map"></div>
<div id="popup" class="ol-popup">
    <div id="popup-content"></div>
</div>
<script type="text/javascript">
    var map = new ol.Map({
        target: 'map',
        layers: [new ol.layer.Tile({
            source: new ol.source.TileWMS({
                url: 'https://egisapp.me.go.kr/geoserver/gwc/service/wms?',
                params: {
                    'LAYERS': 'EGIS:lv3_2023y',
                    'TILED': true,
                    'SRS': 'EPSG:3857'
                }
            })
        })],
        view: new ol.View({
            center: ol.proj.transform([127.88, 37.69], 'EPSG:4326', 'EPSG:3857'),
            zoom: 12,
            enableRotation: false
        })
    });

    var container = document.getElementById('popup');
    var content = document.getElementById('popup-content');
    var overlay = new ol.Overlay({
        element: container,
        autoPan: true,
        autoPanAnimation: {
            duration: 250
        }
    });
    map.addOverlay(overlay);

    var features = [];
"""

# 데이터프레임에서 위도와 경도 데이터를 추출하여 JavaScript 변수에 추가
for idx, row in df.iterrows():
    latitude = row['위도']
    longitude = row['경도']
    branch_name = row['지사명']
    route_name = row['노선명']
    section = row['구간']
    if pd.notna(latitude) and pd.notna(longitude):
        html_content += f"""
    var feature = new ol.Feature({{
        geometry: new ol.geom.Point(ol.proj.fromLonLat([{longitude}, {latitude}])),
        name: '{branch_name}',
        route: '{route_name}',
        section: '{section}'
    }});
    features.push(feature);
"""

# JavaScript 코드 마무리 및 스타일 추가
html_content += """
    var vectorSource = new ol.source.Vector({
        features: features
    });

    var vectorLayer = new ol.layer.Vector({
        source: vectorSource,
        style: new ol.style.Style({
            image: new ol.style.Circle({
                radius: 5,
                fill: new ol.style.Fill({ color: 'red' }),
                stroke: new ol.style.Stroke({ color: 'black', width: 1 })
            })
        })
    });

    map.addLayer(vectorLayer);

    map.on('singleclick', function(evt) {
        var coordinate = evt.coordinate;
        var feature = map.forEachFeatureAtPixel(evt.pixel, function(feature) {
            return feature;
        });
        if (feature) {
            var geometry = feature.getGeometry();
            var coord = geometry.getCoordinates();
            var name = feature.get('name');
            var route = feature.get('route');
            var section = feature.get('section');
            content.innerHTML = '<b>지사명:</b> ' + name + '<br><b>노선명:</b> ' + route + '<br><b>구간:</b> ' + section;
            overlay.setPosition(coord);
        }
    });
</script>
</body>
</html>
"""

# HTML 파일로 저장
with open('map2.html', 'w', encoding='utf-8') as file:
    file.write(html_content)

print("HTML 파일이 생성되었습니다: map2.html")
```

```python
import requests
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops
from PIL import Image
import io

# WMS 서비스 URL 및 레이어 정보 설정
url = 'https://egisapp.me.go.kr/geoserver/gwc/service/wms?'
layer_name = 'EGIS:lv3_2023y'  # 레이어명
srs = 'EPSG:3857'
format = 'image/png'
width = 1024  # 이미지 너비
height = 1024  # 이미지 높이

# 홍천 지역 경계 설정 (EPSG:3857 좌표계 사용)
bbox = '14126252.177964,4468503.300468,14151745.013401,4490940.967702'  

# WMS 요청 매개변수 설정
params = {
    'service': 'WMS',
    'version': '1.1.1',
    'request': 'GetMap',
    'layers': layer_name,
    'styles': '',
    'srs': srs,
    'bbox': bbox,
    'width': width,
    'height': height,
    'format': format,
    'tiled': 'true'
}

# WMS 요청
response = requests.get(url, params=params)

# 응답 내용 확인 및 이미지 시각화
if response.headers['Content-Type'] == 'image/png':
    # 응답 이미지를 PIL 형식으로 변환
    image = Image.open(io.BytesIO(response.content))
    image_np = np.array(image)
    
    # 생물 파편화 식별을 위한 임계값 설정 (예: 숲을 나타내는 값)
    forest_value = 1  # 실제 데이터의 숲 값을 설정해야 합니다.
    
    # 특정 서식지 유형 필터링 (예: 숲)
    forest_mask = (image_np[:, :, 0] == forest_value)
    
    # 생물 파편화 패치 식별
    labeled_forest, num_patches = label(forest_mask, return_num=True, connectivity=2)
    
    # 패치 수 출력
    print("패치 수:", num_patches)
    
    # 생물 파편화 시각화
    plt.imshow(labeled_forest, cmap='nipy_spectral')
    plt.colorbar()
    plt.title('생물 파편화')
    plt.show()
    
    # 패치 면적의 히스토그램
    patch_areas = [region.area for region in regionprops(labeled_forest)]
    plt.hist(patch_areas, bins=30)
    plt.title('패치 면적 분포')
    plt.xlabel('면적')
    plt.ylabel('패치 수')
    plt.show()
    
else:
    print("응답이 이미지가 아닙니다.")
    print("응답 내용:", response.text)
```

```python
import requests
import numpy as np
from skimage.measure import label, regionprops
import matplotlib.pyplot as plt

# WMS 서비스 URL 및 레이어 정보 설정
url = 'http://egisapp.me.go.kr/geoserver/wms?'
layer_name = 'EGIS:lv3_2023y'  # 레이어명
srs = 'EPSG:3857'
format = 'image/png'
width = 256
height = 256

# 홍천 지역 경계 설정 (EPSG:3857 좌표계 사용)
bbox = '14126252.177964,4468503.300468,14151745.013401,4490940.967702'  

# WMS 요청 매개변수 설정
params = {
    'service': 'WMS',
    'version': '1.1.1',
    'request': 'GetMap',
    'layers': layer_name,
    'styles': '',
    'srs': srs,
    'bbox': bbox,
    'width': width,
    'height': height,
    'format': format,
    'tiled': 'true'
}

# WMS 요청
response = requests.get(url, params=params)

# 응답 확인
if response.status_code == 200:
    # 응답 데이터를 numpy 배열로 변환
    image = plt.imread(io.BytesIO(response.content))
    
    # 생물 파편화 식별을 위한 임계값 설정 (예: 숲을 나타내는 값)
    forest_value = 1
    
    # 특정 서식지 유형 필터링 (예: 숲)
    forest_mask = (image[:, :, 0] == forest_value)
    
    # 생물 파편화 패치 식별
    labeled_forest, num_patches = label(forest_mask, return_num=True, connectivity=2)
    
    # 패치 수 출력
    print("패치 수:", num_patches)
    
    # 생물 파편화 시각화
    plt.imshow(labeled_forest, cmap='nipy_spectral')
    plt.colorbar()
    plt.title('생물 파편화')
    plt.show()
    
else:
    print("WMS 요청이 실패했습니다. 응답 코드:", response.status_code)
```
