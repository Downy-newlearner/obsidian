```python
import pandas as pd

df2 = pd.read_csv('./data/전국생태통로.csv')
df2.drop(columns='번호', inplace=True)
df2.head()
```

```python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구의 반경 (km)
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2) ** 2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance
```

```python
before_installation = []
before_year = []
after_installation = []
after_year = []

for idx2, row2 in df2.iterrows():
    for idx1, row1 in df1.iterrows():
        distance = haversine(row2['위도'], row2['경도'], row1['위도'], row1['경도'])
        if distance <= 5:
            if row1['Year'] < row2['설치년도']:
                before_installation.append(row1)
                before_year.append(row2['설치년도'])
            else:
                after_installation.append(row1)
                after_year.append(row2['설치년도'])

before_installation_df = pd.DataFrame(before_installation, before_year)
after_installation_df = pd.DataFrame(after_installation, after_year)
```

```python
# 생태통로 설치 전 데이터
before_installation_df['생태통로설치'] = before_year
before_installation_df
```

```python
before_result = before_installation_df.groupby(['경도', '위도']).agg({
    '국립공원명': 'first', 
    'Year': 'first', 
    '본부명': 'first', 
    '지사명': 'first', 
    '노선명': 'first', 
    '방향': 'first', 
    '발생건수': 'mean', 
    '생태통로설치': 'first'
}).reset_index()

before_result
```

```python
# 생태통로 설치 후 데이터
after_installation_df['생태통로설치'] = after_year
after_installation_df
```

```python
after_result = after_installation_df.groupby(['경도', '위도']).agg({
    '국립공원명': 'first', 
    'Year': 'first', 
    '본부명': 'first', 
    '지사명': 'first', 
    '노선명': 'first', 
    '방향': 'first', 
    '발생건수': 'mean', 
    '생태통로설치': 'first'
}).reset_index()

after_result
```

```python
merged_df = pd.merge(before_result, after_result, on=['위도', '경도'], suffixes=('_before', '_after'))
merged_df['발생건수차이'] = merged_df['발생건수_after'] - merged_df['발생건수_before']
merged_df.head()
```

```python
result_df = merged_df[['국립공원명_before', '생태통로설치_before', '경도', '위도', '발생건수_before', '발생건수_after', '발생건수차이']]
result_df.columns = ['국립공원명', '생태통로설치', '경도', '위도', '설치년도_이전_발생건수', '설치년도_이후_발생건수', '발생건수차이']

result_df
```
