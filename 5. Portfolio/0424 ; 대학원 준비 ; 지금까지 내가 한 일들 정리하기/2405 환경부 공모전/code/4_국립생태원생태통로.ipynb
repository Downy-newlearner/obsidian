```python
import pandas as pd

df['위도_도'] = df['위도_도'].astype(float)
df['위도_분'] = df['위도_분'].astype(float) 
df['위도_초'] = df['위도_초'].astype(float)  
df['경도_도'] = df['경도_도'].astype(float)
df['경도_분'] = df['경도_분'].astype(float)  
df['경도_초'] = df['경도_초'].astype(float) 

df['위도'] = df['위도_도'] + df['위도_분'] / 60 + df['위도_초'] / 3600
df['경도'] = df['경도_도'] + df['경도_분'] / 60 + df['경도_초'] / 3600

df['설치년도'] = df['설치년도'].astype(str).str.extract(r'(\d{4})')

# Drop unnecessary columns
df.drop(columns=['위도_도', '위도_분', '위도_초', '경도_도', '경도_분', '경도_초'], inplace=True)

df.head()
```

### 위도, 경도

```python
import folium
from folium.plugins import HeatMap

korea = folium.Map(location=[36.55, 127.75], zoom_start=7)
HeatMap(data=df[['위도', '경도']], radius=10).add_to(korea)

korea
```

```python
import folium

# 작은 아이콘 이미지 경로
icon_url_red = 'https://raw.githubusercontent.com/pointhi/leaflet-color-markers/master/img/marker-icon-2x-red.png'

# 지도의 초기 위치와 확대 수준 설정
korea = folium.Map(location=[36.55, 127.75], zoom_start=7)

# df 데이터프레임에 있는 각 위치에 작은 빨간색 마커 추가
for idx, row in df.iterrows():
    icon = folium.CustomIcon(
        icon_url_red,
        icon_size=(10, 10)  # 아이콘의 크기 설정
    )
    marker = folium.Marker(
        location=[row['위도'], row['경도']],
        icon=icon
    )
    marker.add_to(korea)

# 지도 표시
korea
```

```python
import folium
from folium.plugins import MarkerCluster

# 지도의 초기 위치와 확대 수준 설정
korea = folium.Map(location=[36.55, 127.75], zoom_start=7)

# 마커 클러스터 생성
mc = MarkerCluster()

# df 데이터프레임의 각 행에 대해 마커 추가
for _, row in df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['주소'] + '\n' + row['생태통로유형'] + '\n' + row['관리기관'] + '\n' + row['설치기관'], 
        icon=folium.Icon(color='red')  # 빨간색 마커
    ).add_to(mc)

# 마커 클러스터를 지도에 추가
korea.add_child(mc)

# 지도 표시
korea
```
